```python
import pandas
import operator
import numpy as np
```

```python
df= pandas.read_csv('../data/lable_stammed.csv')
```

```python
bad_words = open('../data/bad_words_seed.txt').read().split('\n')[:-1]
```

```python
df_bad = df[df['label'] == True]
```

```python
len(df_bad)
```

```python
words = []
for sentence in df_bad.text:
    for word in sentence.split(' '):
        words.append(word)
```

```python
len(words)
```

```python
words =list(filter(lambda a: a != '', words))
```

```python
word_dict = dict()
for word in words:
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1   
```

```python
len(word_dict)
```

```python
count = 0
actual_bad_words = []
for word in bad_words:
    if word in word_dict:
        actual_bad_words.append(word)
        count += 1
```

```python
def coocurance(x, y):
    count = 0
    for sentence in df_bad.text:
        if x in sentence.split(' ') and y in sentence.split(' '):
            count += 1
    return count
```

```python
def check(bad_word):
    for word in word_dict:
        if bad_word != word:
            res.append([bad_word, word, coocurance(bad_word, word)])
```

```python
res = dict()
for sentence in df_bad.text:
    selection = []
    for word in sentence.split(' '):
        if word in actual_bad_words and word_dict[word] >= 5:
            selection.append(word)
    for bad_word in selection:
        for word in sentence.split(' '):
            if bad_word != word and word != '':
                key = bad_word + '-' + word
                if key in res:
                    res[key] += 1
                else:
                    res[key] = 1
        
```
