```python
import pandas
import operator
import numpy as np
from nltk import ngrams
import seaborn as sns
from nltk.stem.snowball import SnowballStemmer
%matplotlib inline
```

```python
df = pandas.read_csv('../data/lable_stammed.csv')
```

```python

bad_words = open('../data/bad_words_seed.txt').read().split('\n')[:-1]
```

```python
df = df.dropna()
```

```python
stemmer = SnowballStemmer("russian", ignore_stopwords=True) 
```

```python
bad_stemmed = dict()
for word in bad_words:
    bad_stemmed[word] = stemmer.stem(word)
```

```python
bad_stemmed
```

```python
all_stemmed = dict()
for sentence in df.text:
    for word in set(sentence.split(' ')):
        if word != '':
            if word not in all_stemmed:
                all_stemmed[word] = stemmer.stem(word)
    
```

```python
all_counts = dict()
for sentence in df.text:
    for word in set(sentence.split(' ')):
        if word != '':
            word = stemmer.stem(word)
            if word in all_counts:
                all_counts[word] += 1
            else:
                all_counts[word] = 1
```

```python
p_bad = dict()
part = df[df.label == True]
for sentence in part.text:
    for word in set(sentence.split(' ')):
        if word != '':
            stemed = stemmer.stem(word)
            if stemed != '':
                if all_counts[stemed] >= 20:
                    if stemed in p_bad:
                        p_bad[stemed] += 1.0/len(part)
                    else:
                        p_bad[stemed] = 1.0/len(part)
```

```python
sort(p_bad)
```

```python
p_good = dict()
part = df[df.label == False]
for sentence in part.text:
    for word in set(sentence.split(' ')):
        if word != '':
            stemed = stemmer.stem(word)
            if stemed != '':
                if all_counts[stemed] >= 20:
                    if stemed in p_good:
                        p_good[stemed] += 1.0/len(part)
                    else:
                        p_good[stemed] = 1.0/len(part)
```

```python
difference = dict()
for key in p_good:
    if key in p_bad:
        difference[key] =(p_bad[key] - p_good[key])/ np.maximum( p_bad[key], p_good[key]) 
```

```python
def sort(x, rev = True):
    return sorted(x.items(), key=operator.itemgetter(1), reverse=rev)
```

```python
sort(difference)
```
