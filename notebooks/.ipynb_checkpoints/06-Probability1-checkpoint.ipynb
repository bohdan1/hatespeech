```python
import pandas
import operator
import numpy as np
from nltk import ngrams
```

```python
df = pandas.read_csv('../data/lable.csv')
```

```python
unique_words = set()
for sentence in df.text:
    for word in sentence.split(' '):
        if word != '':
            unique_words.add(word)
        
    
```

```python
probability = dict()
for sentence in df.text:
    for word in set(sentence.split(' ')):
        if word != '':
            if word in probability:
                probability[word] += 1.0/len(df)
            else:
                probability[word] = 1.0/len(df)
```

```python
def sort(x):
    return sorted(x.items(), key=operator.itemgetter(1), reverse=True)
```

```python
sort(probability)
```

```python
p_bad = dict()
bad = df[df.label == True]
for sentence in bad.text:
    for word in set(sentence.split(' ')):
        if word != '':
            if word in p_bad:
                p_bad[word] += 1.0/len(bad)
            else:
                p_bad[word] = 1.0/len(bad)
```

```python
sort(p_bad)
```

```python
p_good = dict()
bad = df[df.label == False]
for sentence in bad.text:
    for word in set(sentence.split(' ')):
        if word != '':
            if word in p_good:
                p_good[word] += 1.0/len(bad)
            else:
                p_good[word] = 1.0/len(bad)
```

```python
sort(p_good)
```

```python
difference = dict()
for key in p_good:
    if key in p_bad:
        difference[key] =(p_bad[key] - p_good[key])/ np.maximum( p_bad[key], p_good[key])
```
