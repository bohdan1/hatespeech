```python
df_bad = df[df['label'] == True]
```

```python
words = []
for sentence in df_bad.text:
    for word in sentence.split(' '):
        words.append(word)
```

```python
len(words)
```

```python
words =list(filter(lambda a: a != '', words))
```

```python
word_dict = dict()
for word in words:
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1   
```

```python
len(word_dict)
```

```python
count = 0
actual_bad_words = []
for word in bad_words:
    if word in word_dict:
        actual_bad_words.append(word)
        count += 1
```
