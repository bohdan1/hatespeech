```python
import pandas
import operator
```

```python
df= pandas.read_csv('../data/lable.csv')
```

```python
bad_part = []
```

```python
for sentence in df[df['label'] == True].text:
    for word in sentence.split(' '):
        bad_part.append(word)
```

```python
for i in range(len(bad_part)):
    bad_part[i]= bad_part[i].replace('\ufeff', '')
    bad_part[i]= bad_part[i].replace('\r\n', '')
    bad_part[i]= bad_part[i].replace('\xa0', '')
```

```python
bad_part =list(filter(lambda a: a != '', bad_part))
```

```python
len(bad_part)
```

```python
bad_words = dict()
```

```python
for word in bad_part:
    if word in bad_words:
        bad_words[word] += 1
    else:
        bad_words[word] = 1   
```

```python
len(bad_words)
```

```python
sb = sorted(bad_words.items(), key=operator.itemgetter(1), reverse=True)
```

```python
for key in bad_words:
    bad_words[key] /=  len(bad_words)
```

```python
bad_words
```

```python
good_part = []
```

```python
good_words = dict()
```

```python
for key in good_words:
    good_words[key] /=  len(good_words)
```
