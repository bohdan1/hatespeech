```python
import pandas
import operator
import numpy as np
from nltk import ngrams
import seaborn as sns
%matplotlib inline
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("russian", ignore_stopwords=True) 
```

```python
df = pandas.read_csv('../data/lable.csv')
```

```python
unique_words = set()
for sentence in df.text:
    for word in sentence.split(' '):
        if word != '':
            word = stemmer.stem(word)
            unique_words.add(word)
        
    
```

```python
probability = dict()
for sentence in df.text:
    for word in set(sentence.split(' ')):
        if word != '':
            word = stemmer.stem(word)
            if word in probability:
                probability[word] += 1.0
            else:
                probability[word] = 1.0
```

```python
sort(probability, False)
```

```python
def sort(x, rev = True):
    return sorted(x.items(), key=operator.itemgetter(1), reverse=rev)
```

```python
counts = []
for key in probability:
    counts.append(probability[key])
```

```python
p_bad = dict()
bad = df[df.label == True]
for sentence in bad.text:
    for word in set(sentence.split(' ')):
        if word != '':
            word = stemmer.stem(word)
            if word in p_bad:
                p_bad[word] += 1.0/len(bad)
            else:
                p_bad[word] = 1.0/len(bad)
```

```python
sort(p_bad)
```

```python
p_good = dict()
bad = df[df.label == False]
for sentence in bad.text:
    for word in set(sentence.split(' ')):
        if word != '':
            word = stemmer.stem(word)
            if word in p_good:
                p_good[word] += 1.0/len(bad)
            else:
                p_good[word] = 1.0/len(bad)
```

```python
sort(p_good)
```

```python
difference = dict()
for key in p_good:
    if key in p_bad:
        difference[key] =(p_bad[key] - p_good[key])/ np.maximum( p_bad[key], p_good[key]) 
```

```python
sort(difference, True)
```
